## Object Oriented APIs 

```python
# First Method 

fig = plt.figure() # creates a figure 
ax = fig.add_subplot() # adds some axes 
plt.show() 
```

```python
# Second Method 

fig = plt.figure() # creates a figure 
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y); # add some data 
plt.show(x,y) 
```

```python
# Third Method (recommended)

fig, ax = plt.subplots()
ax.plot(x, y); # Adds some data 
```

## Matplotlib; example workflows 

```python
# 0. Import matplotlib and get ready for plotting in Jupyter 
%matplotlib inline
import matplotlib.pyplot as plt 

# 1. Prepare the data 
x = [1, 2, 33, 4, 5]         # 5 elements
y = [4,5,6,7,8]      # Now 5 elements

# 2. Setup plot 
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot the data 
ax.plot(x, y)

# 4. Customize the plot 
ax.set(title="Sample Plot", 
       xlabel="x-axis", 
       ylabel="y-axis")

# 5. Save & Show (save the whole figure locally)
fig.savefig("sample-plot.png")
plt.show()  # This will display the plot in Jupyter
```

## Making figures with; Numpy array 

Here we will be plotting:
1. Line
2. Scatter
3. Bar
4. Histogran
5. Subpots 

```python
import numpy as np 
```

```python
# Create some data 

x = np.linspace(0,10,100)
x[:10]
```

```python
# Line plot 

fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Scatter plot 

fig, ax = plt.subplots()
ax.scatter(x, np.square(x**2))
```

```python
# Bar Plot 

a = ["Saurabh", "Ranjan", "Singh", "Suryavanshi"]
b = [20, 30, 40, 50]

fig, ax = plt.subplots()
ax.bar(a,b)
ax.set(title="Random Number", 
      xlabel="Name", 
      ylabel="Number")
```

```python
# Histogram plot 

data = np.random.randn(10000)

fig, ax = plt.subplots()
ax.hist(data)
```

```python
# Generate random data
data1 = np.random.randn(10000)
data2 = np.random.randn(10000)

# Create two subplots side by side
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))

# Plot a histogram on each subplot
ax1.hist(data1)
ax1.set(title="Histogram 1")

ax2.hist(data2)
ax2.set(title="Histogram 2")

plt.show()
```
